## üìñ About the Dataset

The **Wind & Solar Energy Production Dataset** contains hourly wind and solar generation data from France spanning **January 2020 to November 2025**, featuring **51,864 complete records** with 9 key columns.

This dataset includes temporal features (date, hours, day-of-year, day name, month, season) and source classification (Wind, Solar, Mixed), with total production ranging from **58 to 23,446 MWh** per hour and wind dominating at **81.9%** of records.

### üåü Key Highlights

- **‚è±Ô∏è Hourly Granularity:** Detailed hour-by-hour energy production measurements
- **üìÖ 5+ Years Coverage:** Comprehensive data from 2020-2025
- **‚úÖ 100% Complete:** Zero missing values across all features
- **üîã Dual Sources:** Wind and Solar energy production tracking
- **üßÆ Rich Features:** 9 engineered features for advanced analysis

## üîç Data Quality Assessment

### Checking for Missing Values and Duplicates

```python
# Check for missing values
missing_values = df.isnull().sum()
total_missing = missing_values.sum()

print("üìä Missing Values Analysis:")
print("="*60)
print(f"\n{'Column':<20} {'Missing Count':<15} {'Percentage'}")
print("-"*60)

for col in df.columns:
    miss_count = missing_values[col]
    miss_pct = (miss_count / len(df)) * 100
    print(f"{col:<20} {miss_count:<15} {miss_pct:.2f}%")

print("-"*60)
print(f"\n‚úÖ Total Missing Values: {total_missing}")
print(f"‚úÖ Data Completeness: {((1 - total_missing/(len(df)*len(df.columns)))*100):.2f}%")
```

```python
# Check for duplicate records
duplicate_count = df.duplicated().sum()

print("üîç Duplicate Records Analysis:")
print("="*60)
print(f"\n‚úÖ Total Records: {len(df):,}")
print(f"‚úÖ Duplicate Records: {duplicate_count:,}")
print(f"‚úÖ Unique Records: {len(df) - duplicate_count:,}")
print(f"‚úÖ Duplicate Percentage: {(duplicate_count/len(df)*100):.2f}%")

if duplicate_count == 0:
    print("\nüéâ Excellent! No duplicate records found in the dataset.")
```

---

## üîß Feature Engineering & Data Preprocessing

```python
# Convert Date column to datetime and extract Year
df["Date"] = pd.to_datetime(df["Date"])
df["Year"] = df["Date"].dt.year

# Remove Mixed records (only 2 out of 51,864)
print(f"üìä Original Dataset Shape: {df.shape}")
print(f"üìä Source Distribution Before Filtering:\n{df['Source'].value_counts()}\n")

df = df[df['Source'] != 'Mixed']

print(f"‚úÖ Dataset After Filtering: {df.shape}")
print(f"‚úÖ Remaining Sources:\n{df['Source'].value_counts()}")
print(f"\nüéØ Mixed records removed for cleaner analysis")
```

```python
# Categorical features distribution
cols_for_count = ["Source", "Day_Name", "Month_Name", "Season", "Year"]

palette_map = {
    "Source": "Set2",        
    "Day_Name": "Pastel1",   
    "Month_Name": "tab20",   
    "Season": "Set3",       
    "Year": "Blues"      
}

plt.figure(figsize=(16, 12))

for i, col in enumerate(cols_for_count):
    plt.subplot(3, 2, i+1)
    
    # Create countplot
    ax = sns.countplot(
        x=col,
        palette=palette_map[col],
        data=df,
        edgecolor='black',
        linewidth=1.5
    )
    
    # Add value labels on bars
    for container in ax.containers:
        ax.bar_label(container, fontsize=9, fontweight='bold')
    
    plt.title(f"Distribution of {col}", fontsize=14, fontweight='bold', pad=10)
    plt.xlabel("")
    plt.ylabel("Count", fontsize=11, fontweight='bold')
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', alpha=0.3, linestyle='--', linewidth=0.7)

plt.tight_layout()
plt.show()

print("üìä **Key Observations:**")
print("   ‚Ä¢ All categorical features show balanced distributions")
print("   ‚Ä¢ Wind energy dominates the Source category")
print("   ‚Ä¢ Seasonal and monthly patterns are evenly represented")
```

```python
# Energy Source Distribution - Pie Chart
source_counts = df["Source"].value_counts()

colors = {
    "Wind": "#1f77b4",   # Blue
    "Solar": "#ff7f0e",  # Orange
}

pie_colors = [colors[src] for src in source_counts.index]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Pie Chart
wedges, texts, autotexts = ax1.pie(
    source_counts.values,
    labels=source_counts.index,
    autopct="%1.1f%%",
    startangle=90,
    colors=pie_colors,
    wedgeprops={"edgecolor": "black", "linewidth": 2},
    textprops={'fontsize': 12, 'weight': 'bold'}
)

ax1.set_title("Energy Production Source Distribution", fontsize=16, fontweight='bold', pad=15)

# Bar Chart with Total Production
source_production = df.groupby('Source')['Production'].sum().sort_values(ascending=False)
bars = ax2.bar(
    source_production.index,
    source_production.values,
    color=pie_colors,
    edgecolor='black',
    linewidth=2,
    alpha=0.8
)

ax2.set_title('Total Production by Energy Source', fontsize=16, fontweight='bold', pad=15)
ax2.set_ylabel('Total Production (MWh)', fontsize=12, fontweight='bold')
ax2.set_xlabel('Energy Source', fontsize=12, fontweight='bold')
ax2.grid(True, axis='y', alpha=0.3, linestyle='--', linewidth=0.7)

# Add value labels on bars
for bar in bars:
    height = bar.get_height()
    ax2.text(
        bar.get_x() + bar.get_width()/2.,
        height,
        f'{height/1e6:.2f}M',
        ha='center',
        va='bottom',
        fontsize=11,
        fontweight='bold'
    )

plt.tight_layout()
plt.show()

print("\nüí® **Wind Energy Insights:**")
print(f"   ‚Ä¢ Dominates with {source_counts['Wind']:,} records ({source_counts['Wind']/len(df)*100:.1f}%)")
print(f"   ‚Ä¢ Total production: {source_production['Wind']/1e6:.2f} million MWh")
print("\n‚òÄÔ∏è **Solar Energy Insights:**")
print(f"   ‚Ä¢ {source_counts['Solar']:,} records ({source_counts['Solar']/len(df)*100:.1f}%)")
print(f"   ‚Ä¢ Total production: {source_production['Solar']/1e6:.2f} million MWh")
```

```python
# Hourly Production Distribution with Histogram and KDE
fig, ax = plt.subplots(figsize=(14, 7))

# Create histogram with KDE
n, bins, patches = ax.hist(
    df['Production'],
    bins=60,
    color='skyblue',
    edgecolor='black',
    linewidth=1.5,
    alpha=0.7,
    label='Frequency'
)

ax2 = ax.twinx()
df['Production'].plot(
    kind='kde',
    ax=ax2,
    color='orange',
    linewidth=2.5,
    label='KDE'
)

ax2.set_ylabel('Density', fontsize=14, fontweight='bold')
ax2.tick_params(axis='y', labelsize=11)

# Formatting
ax.set_title('Hourly Renewable Energy Production Distribution', 
             fontsize=16, fontweight='bold', pad=15)
ax.set_xlabel('Production (MWh)', fontsize=14, fontweight='bold')
ax.set_ylabel('Frequency', fontsize=14, fontweight='bold')
ax.tick_params(axis='both', labelsize=11)

# Add statistics box
stats_text = (f'Mean: {df["Production"].mean():.0f} MWh\n'
              f'Median: {df["Production"].median():.0f} MWh\n'
              f'Std Dev: {df["Production"].std():.0f} MWh\n'
              f'Min: {df["Production"].min():.0f} MWh\n'
              f'Max: {df["Production"].max():.0f} MWh')

ax.text(0.98, 0.97, stats_text, transform=ax.transAxes, fontsize=11,
        verticalalignment='top', horizontalalignment='right',
        bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8,
                 edgecolor='black', linewidth=1.5))

# Legends
ax.legend(loc='upper left', fontsize=11, frameon=True, edgecolor='black')
ax2.legend(loc='upper right', fontsize=11, frameon=True, edgecolor='black')

ax.grid(True, alpha=0.3, linestyle='--', linewidth=0.7)

plt.tight_layout()
plt.show()

print("üìä **Distribution Insights:**")
print("   ‚Ä¢ Right-skewed distribution with most production between 3,000-8,000 MWh")
print("   ‚Ä¢ High variability indicating diverse production conditions")
print(f"   ‚Ä¢ Range: {df['Production'].min():.0f} to {df['Production'].max():.0f} MWh")
```

```python
# Boxplot - Production by Energy Source
plt.figure(figsize=(10, 7))

ax = sns.boxplot(
    data=df,
    x="Source",
    y="Production",
    palette="Set2",
    linewidth=2,
    flierprops=dict(marker='o', markerfacecolor='red', markersize=5,
                   markeredgecolor='black', alpha=0.5)
)

# Add mean markers
for i, source in enumerate(df['Source'].unique()):
    mean_val = df[df['Source'] == source]['Production'].mean()
    ax.plot(i, mean_val, marker='D', markersize=10, color='green',
           markeredgecolor='black', markeredgewidth=1.5, zorder=3,
           label='Mean' if i == 0 else '')

plt.title("Production Distribution by Energy Source", fontsize=16, fontweight='bold', pad=15)
plt.xlabel("Energy Source", fontsize=14, fontweight='bold')
plt.ylabel("Production (MWh)", fontsize=14, fontweight='bold')
plt.grid(axis='y', alpha=0.3, linestyle='--', linewidth=0.7)
plt.legend(fontsize=11, loc='upper right', frameon=True, edgecolor='black')

plt.tight_layout()
plt.show()

# Calculate statistics
for source in df['Source'].unique():
    source_data = df[df['Source'] == source]['Production']
    print(f"\n{source} Energy Statistics:")
    print(f"   ‚Ä¢ Mean: {source_data.mean():.0f} MWh")
    print(f"   ‚Ä¢ Median: {source_data.median():.0f} MWh")
    print(f"   ‚Ä¢ Std Dev: {source_data.std():.0f} MWh")
    print(f"   ‚Ä¢ IQR: {source_data.quantile(0.75) - source_data.quantile(0.25):.0f} MWh")

print("\nüîç **Key Observations:**")
print("   üí® Wind shows wider spread with higher variability")
print("   ‚òÄÔ∏è Solar displays more concentrated distribution")
print("   ‚ö° Wind has longer upper tail indicating extreme production events")
```

```python
### üìÖ Temporal Pattern Analysis
```

```python
# Monthly Production Trends
month_order = [
    "January", "February", "March", "April", "May", "June",
    "July", "August", "September", "October", "November", "December"
]

df["Month_Name"] = pd.Categorical(
    df["Month_Name"],
    categories=month_order,
    ordered=True
)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Line Plot - Average Monthly Production
monthly_avg = df.groupby('Month_Name')['Production'].mean().reindex(month_order)

ax1.plot(
    range(len(monthly_avg)),
    monthly_avg.values,
    marker="o",
    linewidth=2.5,
    markersize=10,
    color="#E69F00",
    markerfacecolor="black",
    markeredgecolor="black",
    markeredgewidth=1.5
)

ax1.set_title("Average Monthly Energy Production", fontsize=14, fontweight='bold', pad=12)
ax1.set_xlabel("Month", fontsize=12, fontweight='bold')
ax1.set_ylabel("Average Production (MWh)", fontsize=12, fontweight='bold')
ax1.set_xticks(range(len(monthly_avg)))
ax1.set_xticklabels([m[:3] for m in month_order], rotation=45, ha='right')
ax1.grid(True, alpha=0.3, linestyle='--', linewidth=0.7)

# Bar Chart - Monthly Production
colors = plt.cm.YlGnBu(np.linspace(0.3, 0.9, len(monthly_avg)))
bars = ax2.bar(
    range(len(monthly_avg)),
    monthly_avg.values,
    color=colors,
    edgecolor='black',
    linewidth=1.5,
    alpha=0.85
)

ax2.set_title('Monthly Energy Production', fontsize=14, fontweight='bold', pad=12)
ax2.set_xlabel('Month', fontsize=12, fontweight='bold')
ax2.set_ylabel('Average Production (MWh)', fontsize=12, fontweight='bold')
ax2.set_xticks(range(len(monthly_avg)))
ax2.set_xticklabels([m[:3] for m in month_order], rotation=45, ha='right')
ax2.grid(True, axis='y', alpha=0.3, linestyle='--', linewidth=0.7)

# Add value labels
for bar in bars:
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height,
            f'{height:.0f}', ha='center', va='bottom', fontsize=9, fontweight='bold')

plt.tight_layout()
plt.show()

# Find peak and lowest months
peak_month = monthly_avg.idxmax()
lowest_month = monthly_avg.idxmin()

print("üìÜ **Monthly Pattern Insights:**")
print(f"   ‚Ä¢ Peak Production: {peak_month} ({monthly_avg.max():.0f} MWh)")
print(f"   ‚Ä¢ Lowest Production: {lowest_month} ({monthly_avg.min():.0f} MWh)")
print(f"   ‚Ä¢ Variation: {((monthly_avg.max() - monthly_avg.min())/monthly_avg.mean()*100):.1f}%")
print("\n   ‚ùÑÔ∏è Winter‚Äìearly spring shows high production")
print("   ‚òÄÔ∏è Summer months remain moderate")
print("   üçÇ Strong rebound from September onwards")
```

### üìä 2. Time Series Decomposition

```python
# Time Series Decomposition using STL
from statsmodels.tsa.seasonal import seasonal_decompose

# Prepare time series data
df_ts = df.copy()
df_ts = df_ts.sort_values('Date')
df_ts['DateTime'] = pd.to_datetime(df_ts['Date'].astype(str) + ' ' + df_ts['Start_Hour'].astype(str) + ':00:00')
df_ts = df_ts.set_index('DateTime')

# Aggregate to daily for decomposition (hourly has too much noise)
daily_prod = df_ts['Production'].resample('D').mean()

# Perform decomposition
print("üîÑ Performing seasonal decomposition...")
decomposition = seasonal_decompose(daily_prod, model='additive', period=365)

# Plot decomposition
fig, axes = plt.subplots(4, 1, figsize=(16, 12))

# Original
axes[0].plot(decomposition.observed, color='blue', linewidth=1.5)
axes[0].set_ylabel('Original', fontsize=12, fontweight='bold')
axes[0].set_title('Time Series Decomposition (STL)', fontsize=16, fontweight='bold', pad=15)
axes[0].grid(True, alpha=0.3)

# Trend
axes[1].plot(decomposition.trend, color='green', linewidth=2)
axes[1].set_ylabel('Trend', fontsize=12, fontweight='bold')
axes[1].grid(True, alpha=0.3)

# Seasonal
axes[2].plot(decomposition.seasonal, color='orange', linewidth=1)
axes[2].set_ylabel('Seasonal', fontsize=12, fontweight='bold')
axes[2].grid(True, alpha=0.3)

# Residual
axes[3].plot(decomposition.resid, color='red', linewidth=0.5, alpha=0.7)
axes[3].set_ylabel('Residual', fontsize=12, fontweight='bold')
axes[3].set_xlabel('Date', fontsize=12, fontweight='bold')
axes[3].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\nüìä **Decomposition Insights:**")
print(f"   ‚Ä¢ Trend Component: Shows long-term production changes")
print(f"   ‚Ä¢ Seasonal Component: Captures yearly patterns (365-day cycle)")
print(f"   ‚Ä¢ Residual Component: Represents random noise and anomalies")
print(f"   ‚Ä¢ Seasonal Strength: {1 - (decomposition.resid.var() / decomposition.seasonal.var()):.3f}")
print(f"   ‚Ä¢ Trend Strength: {1 - (decomposition.resid.var() / decomposition.trend.var()):.3f}")
```

### üîç 3. Stationarity Testing & ACF/PACF Analysis

```python
# Stationarity Testing
from statsmodels.tsa.stattools import adfuller, kpss

def adf_test(series, name=''):
    """Perform Augmented Dickey-Fuller test"""
    result = adfuller(series.dropna())
    print(f'\nüìä ADF Test Results for {name}:')
    print(f'   ‚Ä¢ ADF Statistic: {result[0]:.6f}')
    print(f'   ‚Ä¢ p-value: {result[1]:.6f}')
    print(f'   ‚Ä¢ Critical Values:')
    for key, value in result[4].items():
        print(f'      - {key}: {value:.3f}')
    
    if result[1] <= 0.05:
        print(f"   ‚úÖ Series is STATIONARY (reject H0)")
    else:
        print(f"   ‚ùå Series is NON-STATIONARY (fail to reject H0)")
    
    return result[1]

def kpss_test(series, name=''):
    """Perform KPSS test"""
    result = kpss(series.dropna(), regression='c', nlags='auto')
    print(f'\nüìä KPSS Test Results for {name}:')
    print(f'   ‚Ä¢ KPSS Statistic: {result[0]:.6f}')
    print(f'   ‚Ä¢ p-value: {result[1]:.6f}')
    print(f'   ‚Ä¢ Critical Values:')
    for key, value in result[3].items():
        print(f'      - {key}: {value:.3f}')
    
    if result[1] >= 0.05:
        print(f"   ‚úÖ Series is STATIONARY (fail to reject H0)")
    else:
        print(f"   ‚ùå Series is NON-STATIONARY (reject H0)")
    
    return result[1]

# Test on daily data
print("="*80)
print("üî¨ STATIONARITY TESTING")
print("="*80)

adf_p = adf_test(daily_prod, 'Daily Production')
kpss_p = kpss_test(daily_prod, 'Daily Production')

# Apply differencing if non-stationary
if adf_p > 0.05:
    print("\nüîÑ Applying First-Order Differencing...")
    daily_prod_diff = daily_prod.diff().dropna()
    adf_test(daily_prod_diff, 'Differenced Daily Production')
    kpss_test(daily_prod_diff, 'Differenced Daily Production')
```

```python
# ACF and PACF Plots
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

fig, axes = plt.subplots(2, 2, figsize=(16, 10))

# ACF - Original
plot_acf(daily_prod.dropna(), lags=50, ax=axes[0, 0], color='blue', alpha=0.05)
axes[0, 0].set_title('Autocorrelation Function (ACF) - Original', fontsize=12, fontweight='bold')
axes[0, 0].grid(True, alpha=0.3)

# PACF - Original
plot_pacf(daily_prod.dropna(), lags=50, ax=axes[0, 1], color='blue', alpha=0.05, method='ywm')
axes[0, 1].set_title('Partial Autocorrelation Function (PACF) - Original', fontsize=12, fontweight='bold')
axes[0, 1].grid(True, alpha=0.3)

# ACF - Differenced
if adf_p > 0.05:
    plot_acf(daily_prod_diff.dropna(), lags=50, ax=axes[1, 0], color='green', alpha=0.05)
    axes[1, 0].set_title('ACF - Differenced Series', fontsize=12, fontweight='bold')
    axes[1, 0].grid(True, alpha=0.3)
    
    # PACF - Differenced
    plot_pacf(daily_prod_diff.dropna(), lags=50, ax=axes[1, 1], color='green', alpha=0.05, method='ywm')
    axes[1, 1].set_title('PACF - Differenced Series', fontsize=12, fontweight='bold')
    axes[1, 1].grid(True, alpha=0.3)
else:
    axes[1, 0].text(0.5, 0.5, 'Series already stationary', ha='center', va='center', fontsize=14)
    axes[1, 1].text(0.5, 0.5, 'No differencing needed', ha='center', va='center', fontsize=14)

plt.tight_layout()
plt.show()

print("\nüìà **ACF/PACF Interpretation:**")
print("   ‚Ä¢ ACF shows correlation with past values at different lags")
print("   ‚Ä¢ PACF shows direct correlation after removing intermediate effects")
print("   ‚Ä¢ Use these to determine ARIMA parameters (p, d, q)")
print("   ‚Ä¢ Seasonal patterns visible if repeating spikes at seasonal lags")
```

### ü§ñ 4. Feature Engineering for Machine Learning

```python
# Advanced Feature Engineering
df_ml = df_ts.copy()

print("üîß Creating Advanced Features for ML Modeling...")

# 1. Lag Features
for lag in [1, 24, 168]:  # 1 hour, 1 day, 1 week
    df_ml[f'production_lag_{lag}'] = df_ml['Production'].shift(lag)

# 2. Rolling Statistics
for window in [3, 6, 24]:  # 3h, 6h, 24h windows
    df_ml[f'production_rolling_mean_{window}'] = df_ml['Production'].rolling(window=window).mean()
    df_ml[f'production_rolling_std_{window}'] = df_ml['Production'].rolling(window=window).std()

# 3. Rate of Change
df_ml['production_diff_1'] = df_ml['Production'].diff(1)
df_ml['production_pct_change_1'] = df_ml['Production'].pct_change(1)

# 4. Cyclical Encoding (sine/cosine for hour and day_of_year)
df_ml['hour_sin'] = np.sin(2 * np.pi * df_ml['Start_Hour'] / 24)
df_ml['hour_cos'] = np.cos(2 * np.pi * df_ml['Start_Hour'] / 24)
df_ml['day_of_year_sin'] = np.sin(2 * np.pi * df_ml['Day_of_Year'] / 365)
df_ml['day_of_year_cos'] = np.cos(2 * np.pi * df_ml['Day_of_Year'] / 365)

# 5. Exponential Moving Averages
df_ml['ema_12'] = df_ml['Production'].ewm(span=12, adjust=False).mean()
df_ml['ema_24'] = df_ml['Production'].ewm(span=24, adjust=False).mean()

# Drop NaN values created by lag and rolling features
df_ml_clean = df_ml.dropna()

print(f"\n‚úÖ Feature Engineering Complete!")
print(f"   ‚Ä¢ Original Features: {len(df.columns)}")
print(f"   ‚Ä¢ Total Features: {len(df_ml_clean.columns)}")
print(f"   ‚Ä¢ New Features Created: {len(df_ml_clean.columns) - len(df.columns)}")
print(f"   ‚Ä¢ Records after cleaning: {len(df_ml_clean):,}")

# Display sample of new features
print("\nüìä Sample of Engineered Features:")
feature_cols = [col for col in df_ml_clean.columns if 'production' in col.lower() or 'sin' in col or 'cos' in col or 'ema' in col]
print(df_ml_clean[feature_cols[:10]].head())
```

```python
# FASTER ALTERNATIVE: Simple ARIMA (without seasonal component)
# NOTE: SARIMA with s=365 is extremely slow (hours/days). Using faster alternatives.

from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import time

print("ü§ñ Building Time Series Model...")
print("="*80)
print("\n‚ö†Ô∏è  NOTE: SARIMA with 365-day seasonality is computationally prohibitive")
print("   ‚Ä¢ Would take hours or days to train")
print("   ‚Ä¢ Using faster ARIMA model instead")
print("   ‚Ä¢ For full seasonal modeling, consider Prophet or weekly aggregation\n")

# Split data (80% train, 20% test)
train_size = int(len(daily_prod) * 0.8)
train_data = daily_prod[:train_size]
test_data = daily_prod[train_size:]

print(f"üìä Data Split:")
print(f"   ‚Ä¢ Total samples: {len(daily_prod)}")
print(f"   ‚Ä¢ Training samples: {len(train_data)} ({len(train_data)/len(daily_prod)*100:.1f}%)")
print(f"   ‚Ä¢ Testing samples: {len(test_data)} ({len(test_data)/len(daily_prod)*100:.1f}%)")
print(f"   ‚Ä¢ Train period: {train_data.index[0].date()} to {train_data.index[-1].date()}")
print(f"   ‚Ä¢ Test period: {test_data.index[0].date()} to {test_data.index[-1].date()}")

# OPTION 1: Simple ARIMA (FAST - ~30 seconds)
order = (1, 1, 1)  # (p, d, q) - no seasonal component
seasonal_order = (0, 0, 0, 0)  # No seasonality for speed

print(f"\nüéØ Model Configuration (Faster Version):")
print(f"   ‚Ä¢ Model Type: ARIMA (non-seasonal)")
print(f"   ‚Ä¢ Order (p,d,q): {order}")
print(f"   ‚Ä¢ Training time: ~30-60 seconds")

# Fit model
start_time = time.time()
print(f"\n‚è≥ Training ARIMA model...")

model = SARIMAX(train_data, 
                order=order,
                seasonal_order=seasonal_order,
                enforce_stationarity=False,
                enforce_invertibility=False)

results = model.fit(disp=False, maxiter=200)

training_time = time.time() - start_time

print(f"\n‚úÖ Model trained successfully in {training_time:.1f} seconds!")
print(f"\nüìä Model Summary:")
print(results.summary())
```

### üìù Why SARIMA(1,1,1)(1,1,1,365) is Extremely Slow

<div style="background: linear-gradient(135deg, #fc466b 0%, #3f5efb 100%); color: #ffffff; padding: 20px 25px; border-radius: 10px; margin: 15px 0;">

**Computational Complexity Explanation:**

1. **Massive Parameter Space**: SARIMA with s=365 creates hundreds of parameters to estimate
2. **Memory Requirements**: State space matrices grow to (365+p+q) √ó (365+p+q)
3. **Training Time**: Can take **hours to days** depending on data size
4. **Numerical Instability**: Large seasonal periods cause convergence issues

**Speed Comparison:**
- ‚ö° ARIMA(1,1,1): ~30 seconds
- üêå SARIMA(1,1,1)(1,1,1,52): ~5-10 minutes
- üê¢ SARIMA(1,1,1)(1,1,1,365): **Several hours or fails**

**Better Alternatives:**
1. **Prophet** - Designed for long seasonality (yearly, weekly)
2. **Weekly Aggregation** - Reduce to weekly data with s=52
3. **TBATS/BATS** - Multiple seasonal periods
4. **Neural Networks** - LSTM with external regressors

</div>
