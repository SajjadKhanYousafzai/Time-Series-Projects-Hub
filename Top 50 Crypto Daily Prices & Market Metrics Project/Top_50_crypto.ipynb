````python
# import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
# load the data
import os
import glob

# Get all CSV files from the Dataset folder
dataset_path = 'Dataset/'
csv_files = glob.glob(os.path.join(dataset_path, '*.csv'))

# Create a dictionary to store all dataframes
crypto_data = {}

# Load each CSV file into the dictionary
for file in csv_files:
    # Extract crypto name from filename (without .csv extension)
    crypto_name = os.path.basename(file).replace('.csv', '')
    # Load the CSV file
    crypto_data[crypto_name] = pd.read_csv(file)
    print(f"Loaded {crypto_name}: {crypto_data[crypto_name].shape}")

print(f"\nTotal cryptocurrencies loaded: {len(crypto_data)}")
````

````python
# Explore the data structure
# Let's examine one cryptocurrency as an example (Bitcoin)
print("Bitcoin Data Sample:")
print(crypto_data['bitcoin'].head())
print("\n" + "="*80 + "\n")

print("Bitcoin Data Info:")
print(crypto_data['bitcoin'].info())
print("\n" + "="*80 + "\n")

print("Bitcoin Data Statistics:")
print(crypto_data['bitcoin'].describe())
print("\n" + "="*80 + "\n")

print("Column Names:")
print(crypto_data['bitcoin'].columns.tolist())
print("\n" + "="*80 + "\n")

print("Check for missing values:")
print(crypto_data['bitcoin'].isnull().sum())
````

````python
# Clean and prepare the data
# The first row contains column names, need to remove it and convert data types

cleaned_crypto_data = {}

for crypto_name, df in crypto_data.items():
    # Create a copy of the dataframe
    cleaned_df = df.copy()
    
    # Remove the first row if it contains the column names
    if cleaned_df.iloc[0]['Date'] != cleaned_df.iloc[0]['Date'] or cleaned_df.iloc[0]['Close'] == 'BTC-USD' or 'USD' in str(cleaned_df.iloc[0]['Close']):
        cleaned_df = cleaned_df.iloc[1:].reset_index(drop=True)
    
    # Convert Date column to datetime
    cleaned_df['Date'] = pd.to_datetime(cleaned_df['Date'])
    
    # Convert numeric columns to float
    numeric_columns = ['Open', 'High', 'Low', 'Close', 'Volume']
    for col in numeric_columns:
        cleaned_df[col] = pd.to_numeric(cleaned_df[col], errors='coerce')
    
    # Sort by date
    cleaned_df = cleaned_df.sort_values('Date').reset_index(drop=True)
    
    # Store cleaned data
    cleaned_crypto_data[crypto_name] = cleaned_df

print("Data cleaning completed!")
print(f"\nExample - Bitcoin cleaned data:")
print(cleaned_crypto_data['bitcoin'].head())
print("\nData types:")
print(cleaned_crypto_data['bitcoin'].dtypes)
print("\nMissing values:")
print(cleaned_crypto_data['bitcoin'].isnull().sum())
````

````python
# Step 1: Visualize price trends for top cryptocurrencies
# Select major cryptocurrencies to visualize
major_cryptos = ['bitcoin', 'ethereum', 'binance_coin', 'solana', 'cardano', 'xrp', 'dogecoin', 'polygon']

# Create a large figure with subplots
fig, axes = plt.subplots(4, 2, figsize=(16, 16))
axes = axes.flatten()

for idx, crypto in enumerate(major_cryptos):
    if crypto in cleaned_crypto_data:
        df = cleaned_crypto_data[crypto]
        axes[idx].plot(df['Date'], df['Close'], linewidth=1.5, color='#1f77b4')
        axes[idx].set_title(f'{crypto.replace("_", " ").title()} Price History', fontsize=12, fontweight='bold')
        axes[idx].set_xlabel('Date', fontsize=10)
        axes[idx].set_ylabel('Price (USD)', fontsize=10)
        axes[idx].grid(True, alpha=0.3)
        axes[idx].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

print("Price trend visualization completed!")
````

````python
# Step 2: Calculate daily returns and additional features
enhanced_crypto_data = {}

for crypto_name, df in cleaned_crypto_data.items():
    enhanced_df = df.copy()
    
    # Calculate daily returns (percentage change)
    enhanced_df['Daily_Return'] = enhanced_df['Close'].pct_change() * 100
    
    # Calculate moving averages
    enhanced_df['MA_7'] = enhanced_df['Close'].rolling(window=7).mean()
    enhanced_df['MA_30'] = enhanced_df['Close'].rolling(window=30).mean()
    enhanced_df['MA_90'] = enhanced_df['Close'].rolling(window=90).mean()
    
    # Calculate volatility (rolling standard deviation of returns)
    enhanced_df['Volatility_30'] = enhanced_df['Daily_Return'].rolling(window=30).std()
    
    # Calculate price range
    enhanced_df['Price_Range'] = enhanced_df['High'] - enhanced_df['Low']
    enhanced_df['Price_Range_Pct'] = (enhanced_df['Price_Range'] / enhanced_df['Close']) * 100
    
    # Store enhanced data
    enhanced_crypto_data[crypto_name] = enhanced_df

print("Feature engineering completed!")
print(f"\nExample - Bitcoin with new features:")
print(enhanced_crypto_data['bitcoin'][['Date', 'Close', 'Daily_Return', 'MA_7', 'MA_30', 'Volatility_30']].tail(10))
````

````python
# Step 3: Visualize price with moving averages for Bitcoin
btc_data = enhanced_crypto_data['bitcoin']

fig, axes = plt.subplots(2, 1, figsize=(16, 10))

# Plot 1: Price with Moving Averages
axes[0].plot(btc_data['Date'], btc_data['Close'], label='Close Price', linewidth=1.5, alpha=0.7)
axes[0].plot(btc_data['Date'], btc_data['MA_7'], label='7-Day MA', linewidth=1.5, linestyle='--')
axes[0].plot(btc_data['Date'], btc_data['MA_30'], label='30-Day MA', linewidth=1.5, linestyle='--')
axes[0].plot(btc_data['Date'], btc_data['MA_90'], label='90-Day MA', linewidth=1.5, linestyle='--')
axes[0].set_title('Bitcoin Price with Moving Averages', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Date', fontsize=12)
axes[0].set_ylabel('Price (USD)', fontsize=12)
axes[0].legend(loc='upper left', fontsize=10)
axes[0].grid(True, alpha=0.3)

# Plot 2: Daily Returns
axes[1].plot(btc_data['Date'], btc_data['Daily_Return'], linewidth=1, alpha=0.7, color='#2ca02c')
axes[1].axhline(y=0, color='red', linestyle='--', linewidth=1)
axes[1].set_title('Bitcoin Daily Returns (%)', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Date', fontsize=12)
axes[1].set_ylabel('Return (%)', fontsize=12)
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("Moving averages and returns visualization completed!")
````

````python
# Step 4: Statistical analysis - Compare performance metrics
selected_cryptos = ['bitcoin', 'ethereum', 'binance_coin', 'solana', 'cardano', 'xrp', 'dogecoin', 'polygon']

# Create a summary dataframe
summary_stats = []

for crypto in selected_cryptos:
    if crypto in enhanced_crypto_data:
        df = enhanced_crypto_data[crypto]
        
        # Calculate metrics
        stats = {
            'Cryptocurrency': crypto.replace('_', ' ').title(),
            'Current Price': df['Close'].iloc[-1],
            'Avg Daily Return (%)': df['Daily_Return'].mean(),
            'Volatility (%)': df['Daily_Return'].std(),
            'Max Price': df['Close'].max(),
            'Min Price': df['Close'].min(),
            'Price Range (%)': ((df['Close'].max() - df['Close'].min()) / df['Close'].min()) * 100,
            'Total Days': len(df)
        }
        summary_stats.append(stats)

summary_df = pd.DataFrame(summary_stats)

print("Performance Metrics Summary:")
print("="*100)
print(summary_df.to_string(index=False))
print("\n" + "="*100)

# Visualize comparison
fig, axes = plt.subplots(2, 2, figsize=(16, 10))

# Plot 1: Average Daily Returns
axes[0, 0].bar(summary_df['Cryptocurrency'], summary_df['Avg Daily Return (%)'], color='steelblue')
axes[0, 0].set_title('Average Daily Return Comparison', fontsize=12, fontweight='bold')
axes[0, 0].set_ylabel('Average Return (%)', fontsize=10)
axes[0, 0].tick_params(axis='x', rotation=45)
axes[0, 0].grid(True, alpha=0.3, axis='y')

# Plot 2: Volatility
axes[0, 1].bar(summary_df['Cryptocurrency'], summary_df['Volatility (%)'], color='coral')
axes[0, 1].set_title('Volatility Comparison', fontsize=12, fontweight='bold')
axes[0, 1].set_ylabel('Volatility (Std Dev %)', fontsize=10)
axes[0, 1].tick_params(axis='x', rotation=45)
axes[0, 1].grid(True, alpha=0.3, axis='y')

# Plot 3: Current Price
axes[1, 0].bar(summary_df['Cryptocurrency'], summary_df['Current Price'], color='green')
axes[1, 0].set_title('Current Price Comparison', fontsize=12, fontweight='bold')
axes[1, 0].set_ylabel('Price (USD)', fontsize=10)
axes[1, 0].tick_params(axis='x', rotation=45)
axes[1, 0].grid(True, alpha=0.3, axis='y')

# Plot 4: Price Range %
axes[1, 1].bar(summary_df['Cryptocurrency'], summary_df['Price Range (%)'], color='purple')
axes[1, 1].set_title('Historical Price Range (%)', fontsize=12, fontweight='bold')
axes[1, 1].set_ylabel('Range (%)', fontsize=10)
axes[1, 1].tick_params(axis='x', rotation=45)
axes[1, 1].grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()
````

````python
# Step 5: Correlation analysis between cryptocurrencies
# Create a dataframe with returns for selected cryptocurrencies

# First, we need to find common dates across all cryptocurrencies
selected_cryptos = ['bitcoin', 'ethereum', 'binance_coin', 'solana', 'cardano', 'xrp', 'dogecoin', 'polygon']

# Create returns dataframe
returns_df = pd.DataFrame()

for crypto in selected_cryptos:
    if crypto in enhanced_crypto_data:
        df = enhanced_crypto_data[crypto][['Date', 'Daily_Return']].copy()
        df = df.rename(columns={'Daily_Return': crypto})
        
        if returns_df.empty:
            returns_df = df
        else:
            returns_df = returns_df.merge(df, on='Date', how='outer')

# Set Date as index and drop NaN values
returns_df = returns_df.set_index('Date')
returns_df = returns_df.dropna()

# Calculate correlation matrix
correlation_matrix = returns_df.corr()

print("Correlation Matrix (Daily Returns):")
print("="*100)
print(correlation_matrix.round(3))
print("\n" + "="*100)

# Visualize correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, 
            annot=True, 
            fmt='.2f', 
            cmap='coolwarm', 
            center=0,
            square=True,
            linewidths=1,
            cbar_kws={"shrink": 0.8})
plt.title('Correlation Matrix of Daily Returns', fontsize=14, fontweight='bold', pad=20)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

print("\nKey Insights:")
print(f"- Highest correlation: {correlation_matrix.unstack().sort_values(ascending=False).drop_duplicates().iloc[1]:.3f}")
print(f"- Lowest correlation: {correlation_matrix.unstack().sort_values(ascending=True).iloc[0]:.3f}")
````

````python
# Step 6: Cumulative returns analysis
# Calculate cumulative returns for comparison

plt.figure(figsize=(16, 8))

for crypto in selected_cryptos:
    if crypto in enhanced_crypto_data:
        df = enhanced_crypto_data[crypto].copy()
        # Calculate cumulative returns (starting from 100)
        df['Cumulative_Return'] = (1 + df['Daily_Return']/100).cumprod() * 100
        plt.plot(df['Date'], df['Cumulative_Return'], label=crypto.replace('_', ' ').title(), linewidth=2)

plt.title('Cumulative Returns Comparison (Base = 100)', fontsize=16, fontweight='bold')
plt.xlabel('Date', fontsize=12)
plt.ylabel('Cumulative Return (Base = 100)', fontsize=12)
plt.legend(loc='best', fontsize=10)
plt.grid(True, alpha=0.3)
plt.yscale('log')  # Log scale for better visualization
plt.tight_layout()
plt.show()

# Calculate final cumulative returns
print("\nCumulative Returns Summary:")
print("="*80)
for crypto in selected_cryptos:
    if crypto in enhanced_crypto_data:
        df = enhanced_crypto_data[crypto].copy()
        df['Cumulative_Return'] = (1 + df['Daily_Return']/100).cumprod() * 100
        final_return = df['Cumulative_Return'].iloc[-1]
        print(f"{crypto.replace('_', ' ').title():20s}: {final_return:,.2f}x")
````

````python
# Step 7: Volatility analysis over time
fig, axes = plt.subplots(4, 2, figsize=(16, 16))
axes = axes.flatten()

for idx, crypto in enumerate(selected_cryptos):
    if crypto in enhanced_crypto_data:
        df = enhanced_crypto_data[crypto]
        axes[idx].plot(df['Date'], df['Volatility_30'], linewidth=1.5, color='#d62728')
        axes[idx].set_title(f'{crypto.replace("_", " ").title()} - 30-Day Rolling Volatility', 
                           fontsize=11, fontweight='bold')
        axes[idx].set_xlabel('Date', fontsize=9)
        axes[idx].set_ylabel('Volatility (%)', fontsize=9)
        axes[idx].grid(True, alpha=0.3)
        axes[idx].tick_params(axis='x', rotation=45)
        
        # Add mean volatility line
        mean_vol = df['Volatility_30'].mean()
        axes[idx].axhline(y=mean_vol, color='blue', linestyle='--', linewidth=1, 
                         label=f'Mean: {mean_vol:.2f}%')
        axes[idx].legend(fontsize=8)

plt.tight_layout()
plt.show()

print("Volatility analysis completed!")
````

````python
# Step 8: Calculate Risk-Adjusted Performance Metrics

def calculate_sharpe_ratio(returns, risk_free_rate=0.02):
    """Calculate Sharpe Ratio (assuming 2% annual risk-free rate)"""
    excess_returns = returns.mean() - (risk_free_rate / 365)
    return (excess_returns / returns.std()) * np.sqrt(365) if returns.std() != 0 else 0

def calculate_max_drawdown(prices):
    """Calculate Maximum Drawdown"""
    cummax = prices.cummax()
    drawdown = (prices - cummax) / cummax
    return drawdown.min() * 100

def calculate_sortino_ratio(returns, risk_free_rate=0.02):
    """Calculate Sortino Ratio (focuses on downside volatility)"""
    excess_returns = returns.mean() - (risk_free_rate / 365)
    downside_returns = returns[returns < 0]
    downside_std = downside_returns.std()
    return (excess_returns / downside_std) * np.sqrt(365) if downside_std != 0 else 0

def calculate_calmar_ratio(returns, prices):
    """Calculate Calmar Ratio (return / max drawdown)"""
    annual_return = (1 + returns.mean() / 100) ** 365 - 1
    max_dd = abs(calculate_max_drawdown(prices) / 100)
    return annual_return / max_dd if max_dd != 0 else 0

def calculate_var(returns, confidence=0.95):
    """Calculate Value at Risk (VaR) at given confidence level"""
    return np.percentile(returns.dropna(), (1 - confidence) * 100)

# Calculate metrics for all selected cryptocurrencies
risk_metrics = []

for crypto in selected_cryptos:
    if crypto in enhanced_crypto_data:
        df = enhanced_crypto_data[crypto].copy()
        returns = df['Daily_Return'].dropna()
        prices = df['Close'].dropna()
        
        metrics = {
            'Cryptocurrency': crypto.replace('_', ' ').title(),
            'Sharpe Ratio': calculate_sharpe_ratio(returns),
            'Sortino Ratio': calculate_sortino_ratio(returns),
            'Max Drawdown (%)': calculate_max_drawdown(prices),
            'Calmar Ratio': calculate_calmar_ratio(returns, prices),
            'VaR 95% (%)': calculate_var(returns, 0.95),
            'VaR 99% (%)': calculate_var(returns, 0.99)
        }
        risk_metrics.append(metrics)

risk_df = pd.DataFrame(risk_metrics)

print("Risk-Adjusted Performance Metrics:")
print("="*120)
print(risk_df.to_string(index=False))
print("\n" + "="*120)
````

````python
# Step 9: Portfolio Optimization - Modern Portfolio Theory
# Calculate optimal portfolio allocation

# Prepare returns matrix for portfolio optimization
portfolio_returns = returns_df.copy()

# Calculate expected returns and covariance matrix
expected_returns = portfolio_returns.mean() * 365  # Annualize
cov_matrix = portfolio_returns.cov() * 365  # Annualize

print("Expected Annual Returns (%):")
print("="*80)
for crypto in expected_returns.index:
    print(f"{crypto.replace('_', ' ').title():20s}: {expected_returns[crypto]:.2f}%")

print("\n" + "="*80)

# Simulate random portfolios
num_portfolios = 10000
results = np.zeros((4, num_portfolios))
np.random.seed(42)

for i in range(num_portfolios):
    # Random weights
    weights = np.random.random(len(expected_returns))
    weights /= np.sum(weights)
    
    # Portfolio return
    portfolio_return = np.sum(weights * expected_returns)
    
    # Portfolio volatility
    portfolio_std = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))
    
    # Sharpe ratio (assuming 2% risk-free rate)
    sharpe_ratio = (portfolio_return - 2) / portfolio_std
    
    results[0,i] = portfolio_return
    results[1,i] = portfolio_std
    results[2,i] = sharpe_ratio
    results[3,i] = i

# Find optimal portfolios
max_sharpe_idx = np.argmax(results[2])
min_vol_idx = np.argmin(results[1])

print("\nüéØ Optimal Portfolio Allocations:")
print("="*80)
print("\n1. Maximum Sharpe Ratio Portfolio (Best Risk-Adjusted Returns):")
print(f"   Expected Return: {results[0, max_sharpe_idx]:.2f}%")
print(f"   Volatility: {results[1, max_sharpe_idx]:.2f}%")
print(f"   Sharpe Ratio: {results[2, max_sharpe_idx]:.3f}")
print("\n   Asset Allocation:")

max_sharpe_weights = np.random.RandomState(42)
for _ in range(int(max_sharpe_idx) + 1):
    weights = max_sharpe_weights.random(len(expected_returns))
    weights /= np.sum(weights)

for idx, crypto in enumerate(expected_returns.index):
    if weights[idx] > 0.01:  # Show only allocations > 1%
        print(f"   - {crypto.replace('_', ' ').title():20s}: {weights[idx]*100:.2f}%")

print("\n2. Minimum Volatility Portfolio (Lowest Risk):")
print(f"   Expected Return: {results[0, min_vol_idx]:.2f}%")
print(f"   Volatility: {results[1, min_vol_idx]:.2f}%")
print(f"   Sharpe Ratio: {results[2, min_vol_idx]:.3f}")
print("\n   Asset Allocation:")

min_vol_rng = np.random.RandomState(42)
for _ in range(int(min_vol_idx) + 1):
    weights_min = min_vol_rng.random(len(expected_returns))
    weights_min /= np.sum(weights_min)

for idx, crypto in enumerate(expected_returns.index):
    if weights_min[idx] > 0.01:  # Show only allocations > 1%
        print(f"   - {crypto.replace('_', ' ').title():20s}: {weights_min[idx]*100:.2f}%")
````
