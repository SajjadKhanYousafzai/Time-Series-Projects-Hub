## STEP 1 â€” Load and Prepare Data ğŸ•

```python

# â”€â”€â”€ Load Dataset â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
import os

# Try both relative paths (notebook-level vs parent-level working dir)
possible_paths = [
    '../Data/AEP_hourly.csv',
    './Data/AEP_hourly.csv',
    r'D:\LLM and AIOps Projects\Time-Series-Projects-Hub\Hourly Energy Consumption Time Serires\Data\AEP_hourly.csv'
]

df_raw = None
for p in possible_paths:
    if os.path.exists(p):
        df_raw = pd.read_csv(p)
        print(f"âœ… Loaded from: {p}")
        break

if df_raw is None:
    raise FileNotFoundError("âŒ Could not find AEP_hourly.csv â€” check the path!")

# â”€â”€â”€ Basic Info â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print(f"\n{'='*60}")
print(f"  RAW DATASET OVERVIEW")
print(f"{'='*60}")
print(f"  Shape         : {df_raw.shape[0]:,} rows Ã— {df_raw.shape[1]} columns")
print(f"  Columns       : {df_raw.columns.tolist()}")
print(f"  Dtypes        :\n{df_raw.dtypes}")
print(f"\n  First 5 rows:\n{df_raw.head()}")

# â”€â”€â”€ Preprocessing â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
df = df_raw.copy()
df['Datetime'] = pd.to_datetime(df['Datetime'])
df.set_index('Datetime', inplace=True)
df.sort_index(inplace=True)

energy_col = df.columns[0]   # 'AEP_MW'

# â”€â”€â”€ Check for duplicates & missing values â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
dup_count  = df.index.duplicated().sum()
null_count = df[energy_col].isnull().sum()

# Remove duplicates (keep first)
df = df[~df.index.duplicated(keep='first')]

# Forward-fill any small gaps
df[energy_col] = df[energy_col].ffill()

print(f"\n{'='*60}")
print(f"  CLEANED DATASET")
print(f"{'='*60}")
print(f"  Energy Column : {energy_col}")
print(f"  Shape         : {df.shape[0]:,} rows")
print(f"  Date Range    : {df.index.min()}  â†’  {df.index.max()}")
print(f"  Total Days    : {(df.index.max() - df.index.min()).days:,}")
print(f"  Duplicates    : {dup_count} (removed)")
print(f"  Missing Values: {null_count} (filled)")
print(f"\n  Descriptive Statistics:")
print(df.describe().round(2))
```

## STEP 2 â€” Visualize the Time Series ğŸ“ˆ

```python

# â”€â”€â”€ VISUALIZE THE TIME SERIES â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
fig, axes = plt.subplots(4, 1, figsize=(18, 16))
fig.suptitle('AEP Hourly Energy Consumption â€” EDA Overview', fontsize=16, fontweight='bold', y=1.01)

# 1. Full time series
axes[0].plot(df.index, df[energy_col], color='steelblue', linewidth=0.4, alpha=0.9)
axes[0].set_title('Full Time Series (All Years)', fontweight='bold')
axes[0].set_ylabel('Energy (MW)')
axes[0].xaxis.set_major_locator(mdates.YearLocator())
axes[0].xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
axes[0].grid(True, alpha=0.3)

# 2. Yearly resampled mean
df_yearly = df[energy_col].resample('YE').mean()
axes[1].bar(df_yearly.index.year, df_yearly.values, color='teal', alpha=0.8, edgecolor='white')
axes[1].set_title('Yearly Average Energy Consumption', fontweight='bold')
axes[1].set_ylabel('Avg Energy (MW)')
axes[1].set_xlabel('Year')
axes[1].grid(True, alpha=0.3, axis='y')

# 3. One full year (most recent complete year)
recent_year = df.index.year.max() - 1
df_year = df[df.index.year == recent_year]
axes[2].fill_between(df_year.index, df_year[energy_col],
                     alpha=0.6, color='darkgreen', label=str(recent_year))
axes[2].plot(df_year.index, df_year[energy_col], color='darkgreen', linewidth=0.6)
axes[2].set_title(f'Annual Pattern â€” {recent_year}', fontweight='bold')
axes[2].set_ylabel('Energy (MW)')
axes[2].xaxis.set_major_locator(mdates.MonthLocator())
axes[2].xaxis.set_major_formatter(mdates.DateFormatter('%b'))
axes[2].grid(True, alpha=0.3)

# 4. One week (hourly granularity)
week_start = df[df.index.year == recent_year].index[0]
week_end   = week_start + pd.Timedelta(days=7)
df_week    = df[(df.index >= week_start) & (df.index < week_end)]
axes[3].plot(df_week.index, df_week[energy_col], color='crimson',
             linewidth=2, marker='o', markersize=3, label='Hourly MW')
axes[3].set_title(f'Weekly Hourly Pattern â€” {week_start.date()} to {week_end.date()}', fontweight='bold')
axes[3].set_ylabel('Energy (MW)')
axes[3].xaxis.set_major_locator(mdates.DayLocator())
axes[3].xaxis.set_major_formatter(mdates.DateFormatter('%a %d'))
axes[3].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# â”€â”€â”€ MONTHLY BOXPLOT â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
df_box = df.copy()
df_box['Month'] = df_box.index.month_name().str[:3]
month_order = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
df_box['Month'] = pd.Categorical(df_box['Month'], categories=month_order, ordered=True)

fig2, ax = plt.subplots(figsize=(16, 5))
df_box.boxplot(column=energy_col, by='Month', ax=ax,
               boxprops=dict(color='steelblue'),
               medianprops=dict(color='crimson', linewidth=2),
               whiskerprops=dict(color='steelblue'),
               capprops=dict(color='steelblue'),
               flierprops=dict(marker='o', color='gray', alpha=0.3, markersize=2))
ax.set_title('Monthly Energy Distribution (Boxplot)', fontweight='bold', fontsize=13)
ax.set_xlabel('Month')
ax.set_ylabel('Energy (MW)')
plt.suptitle('')
plt.tight_layout()
plt.show()

print("âœ… Visualizations complete!")
```

## STEP 3 â€” Feature Engineering ğŸ”¥

```python

# â”€â”€â”€ FEATURE ENGINEERING â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
df_feat = df.copy()

# â”€â”€ Temporal calendar features â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
df_feat['hour']         = df_feat.index.hour
df_feat['day']          = df_feat.index.day
df_feat['month']        = df_feat.index.month
df_feat['year']         = df_feat.index.year
df_feat['day_of_week']  = df_feat.index.dayofweek          # 0=Mon â€¦ 6=Sun
df_feat['day_of_year']  = df_feat.index.dayofyear
df_feat['week_of_year'] = df_feat.index.isocalendar().week.astype(int)
df_feat['quarter']      = df_feat.index.quarter
df_feat['is_weekend']   = (df_feat['day_of_week'] >= 5).astype(int)

# â”€â”€ Peak-hour flag (business hours 8 AM â€“ 8 PM) â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
df_feat['is_peak_hour'] = df_feat['hour'].between(8, 20).astype(int)

# â”€â”€ US Federal Holidays (simple rule-based) â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
try:
    import holidays
    us_holidays = holidays.US(years=df_feat['year'].unique().tolist())
    df_feat['is_holiday'] = df_feat.index.normalize().map(
        lambda d: 1 if d in us_holidays else 0
    ).astype(int)
except ImportError:
    df_feat['is_holiday'] = 0
    print("âš ï¸ 'holidays' package not installed â€” is_holiday set to 0")

# â”€â”€ Cyclical encoding (prevent ordinal bias) â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
df_feat['hour_sin']   = np.sin(2 * np.pi * df_feat['hour']   / 24)
df_feat['hour_cos']   = np.cos(2 * np.pi * df_feat['hour']   / 24)
df_feat['dow_sin']    = np.sin(2 * np.pi * df_feat['day_of_week'] / 7)
df_feat['dow_cos']    = np.cos(2 * np.pi * df_feat['day_of_week'] / 7)
df_feat['month_sin']  = np.sin(2 * np.pi * df_feat['month']  / 12)
df_feat['month_cos']  = np.cos(2 * np.pi * df_feat['month']  / 12)
df_feat['doy_sin']    = np.sin(2 * np.pi * df_feat['day_of_year'] / 365)
df_feat['doy_cos']    = np.cos(2 * np.pi * df_feat['day_of_year'] / 365)

# â”€â”€ Lag features â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
for lag in [1, 2, 3, 6, 12, 24, 48, 168]:       # 168 h = 1 week
    df_feat[f'lag_{lag}h'] = df_feat[energy_col].shift(lag)

# â”€â”€ Rolling statistics (24-h and 7-day windows) â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
for window, label in [(24, '24h'), (48, '48h'), (168, '7d')]:
    rolled = df_feat[energy_col].rolling(window=window, min_periods=1)
    df_feat[f'roll_mean_{label}'] = rolled.mean()
    df_feat[f'roll_std_{label}']  = rolled.std()
    df_feat[f'roll_min_{label}']  = rolled.min()
    df_feat[f'roll_max_{label}']  = rolled.max()

# â”€â”€ Exponential weighted mean â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
df_feat['ewm_mean_24h'] = df_feat[energy_col].ewm(span=24, adjust=False).mean()

# â”€â”€ Differencing features â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
df_feat['diff_1h']  = df_feat[energy_col].diff(1)
df_feat['diff_24h'] = df_feat[energy_col].diff(24)

print("âœ… Feature Engineering Complete!")
print(f"   Total features : {len(df_feat.columns)}")
print(f"\n   Feature list:")

feature_groups = {
    "Calendar"   : ['hour','day','month','year','day_of_week','quarter','is_weekend','is_peak_hour','is_holiday'],
    "Cyclical"   : [c for c in df_feat.columns if 'sin' in c or 'cos' in c],
    "Lags"       : [c for c in df_feat.columns if c.startswith('lag_')],
    "Rolling"    : [c for c in df_feat.columns if c.startswith('roll_') or c.startswith('ewm_')],
    "Diff"       : [c for c in df_feat.columns if c.startswith('diff_')],
}
for grp, cols in feature_groups.items():
    print(f"   {grp:12s}: {cols}")
```

## STEP 5 â€” Outlier Detection & Treatment ğŸ”

```python

# â”€â”€â”€ OUTLIER DETECTION (IQR + Z-Score) â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
ts = df[energy_col].copy()

# IQR Method
Q1, Q3 = ts.quantile(0.25), ts.quantile(0.75)
IQR    = Q3 - Q1
lower_iqr = Q1 - 1.5 * IQR
upper_iqr = Q3 + 1.5 * IQR
outliers_iqr = ts[(ts < lower_iqr) | (ts > upper_iqr)]

# Z-Score Method (|z| > 3)
z_scores     = np.abs(stats.zscore(ts))
outliers_z   = ts[z_scores > 3]

print(f"{'='*60}")
print(f"  OUTLIER ANALYSIS")
print(f"{'='*60}")
print(f"  IQR range     : [{lower_iqr:,.0f}  â€”  {upper_iqr:,.0f}] MW")
print(f"  IQR outliers  : {len(outliers_iqr):,} rows ({len(outliers_iqr)/len(ts)*100:.2f}%)")
print(f"  Z-Score out.  : {len(outliers_z):,} rows ({len(outliers_z)/len(ts)*100:.2f}%)")

# â”€â”€ Visualise â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
fig, axes = plt.subplots(1, 3, figsize=(20, 5))

# Time-series with outliers highlighted
axes[0].plot(ts.index, ts.values, color='steelblue', linewidth=0.4, alpha=0.8, label='Normal')
axes[0].scatter(outliers_iqr.index, outliers_iqr.values,
                color='red', s=8, zorder=5, label=f'IQR outliers ({len(outliers_iqr):,})')
axes[0].set_title('Outliers (IQR Method)', fontweight='bold')
axes[0].set_ylabel('MW'); axes[0].legend(fontsize=9)
axes[0].grid(True, alpha=0.3)

# Boxplot
axes[1].boxplot(ts.dropna(), vert=True,
                boxprops=dict(color='steelblue'),
                medianprops=dict(color='crimson', linewidth=2),
                flierprops=dict(marker='o', color='red', alpha=0.3, markersize=3))
axes[1].set_title('Boxplot â€” Energy Distribution', fontweight='bold')
axes[1].set_ylabel('MW')

# Z-score distribution
axes[2].hist(z_scores, bins=60, color='mediumslateblue', alpha=0.8, edgecolor='white')
axes[2].axvline(3, color='red', linestyle='--', linewidth=2, label='|z|=3 threshold')
axes[2].set_title('Z-Score Distribution', fontweight='bold')
axes[2].set_xlabel('|Z-Score|'); axes[2].set_ylabel('Frequency')
axes[2].legend(); axes[2].grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()

# â”€â”€ Cap outliers (winsorize at 1% / 99%) â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
p01, p99 = ts.quantile(0.01), ts.quantile(0.99)
df[energy_col]      = df[energy_col].clip(lower=p01, upper=p99)
df_feat[energy_col] = df_feat[energy_col].clip(lower=p01, upper=p99)

print(f"\nâœ… Outliers winsorized at 1stâ€“99th percentile: [{p01:,.0f} â€” {p99:,.0f}] MW")
```

## STEP 5 â€” Stationarity Check ğŸ“‰

```python

# â”€â”€â”€ STATIONARITY TESTS â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
ts = df[energy_col].dropna()

def run_stationarity_tests(series, label="Original"):
    print(f"\n{'='*60}")
    print(f"  {label}")
    print(f"{'='*60}")

    # ADF Test (H0: unit root / non-stationary)
    adf = adfuller(series, autolag='AIC')
    print(f"\n  ğŸ“Š Augmented Dickey-Fuller (ADF)")
    print(f"     Statistic : {adf[0]:.4f}")
    print(f"     P-value   : {adf[1]:.6f}  {'âœ… STATIONARY' if adf[1]<=0.05 else 'âš ï¸ NON-STATIONARY'}")
    print(f"     Crit. Val : 1%={adf[4]['1%']:.3f}  5%={adf[4]['5%']:.3f}  10%={adf[4]['10%']:.3f}")

    # KPSS Test (H0: stationary)
    kpss_res = kpss(series, regression='c', nlags='auto')
    print(f"\n  ğŸ“Š KPSS Test")
    print(f"     Statistic : {kpss_res[0]:.4f}")
    print(f"     P-value   : {kpss_res[1]:.6f}  {'âœ… STATIONARY' if kpss_res[1]>=0.05 else 'âš ï¸ NON-STATIONARY'}")
    print(f"     Crit. Val : {kpss_res[3]}")

# Test original series
run_stationarity_tests(ts, "Original Series")

# Test 1st-order difference
ts_diff = ts.diff(1).dropna()
run_stationarity_tests(ts_diff, "1st-Order Difference")

# â”€â”€ Rolling statistics plot â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
window = 24 * 7   # 1 week rolling
roll_mean = ts.rolling(window=window).mean()
roll_std  = ts.rolling(window=window).std()

fig, axes = plt.subplots(2, 1, figsize=(16, 8))

axes[0].plot(ts.index, ts.values, color='steelblue', linewidth=0.5, alpha=0.7, label='Original')
axes[0].plot(roll_mean.index, roll_mean.values, color='red', linewidth=1.5, label='Rolling Mean (1W)')
axes[0].plot(roll_std.index, roll_std.values, color='orange', linewidth=1.5, label='Rolling Std (1W)')
axes[0].set_title('Rolling Mean & Std â€” Original Series', fontweight='bold')
axes[0].set_ylabel('MW'); axes[0].legend(); axes[0].grid(True, alpha=0.3)

axes[1].plot(ts_diff.index, ts_diff.values, color='darkgreen', linewidth=0.5, alpha=0.7)
axes[1].set_title('1st-Order Differenced Series', fontweight='bold')
axes[1].set_ylabel('Diff MW'); axes[1].grid(True, alpha=0.3)
axes[1].axhline(0, color='black', linestyle='--', linewidth=1)

plt.tight_layout()
plt.show()
```

## STEP 6 â€” Autocorrelation Analysis ğŸ”

```python

# â”€â”€â”€ ACF & PACF ANALYSIS â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
fig, axes = plt.subplots(2, 2, figsize=(18, 10))
fig.suptitle('Autocorrelation Analysis', fontsize=14, fontweight='bold')

# Original series â€” ACF & PACF
plot_acf(ts,  lags=200, ax=axes[0, 0], alpha=0.05, color='steelblue')
axes[0, 0].set_title('ACF â€” Original Series', fontweight='bold')
axes[0, 0].set_xlabel('Lag (hours)')

plot_pacf(ts, lags=60, ax=axes[0, 1], alpha=0.05, color='steelblue', method='ywm')
axes[0, 1].set_title('PACF â€” Original Series', fontweight='bold')
axes[0, 1].set_xlabel('Lag (hours)')

# 1st-difference â€” ACF & PACF
ts_diff = ts.diff(1).dropna()
plot_acf(ts_diff,  lags=200, ax=axes[1, 0], alpha=0.05, color='darkgreen')
axes[1, 0].set_title('ACF â€” 1st-Order Differenced Series', fontweight='bold')
axes[1, 0].set_xlabel('Lag (hours)')

plot_pacf(ts_diff, lags=60, ax=axes[1, 1], alpha=0.05, color='darkgreen', method='ywm')
axes[1, 1].set_title('PACF â€” 1st-Order Differenced Series', fontweight='bold')
axes[1, 1].set_xlabel('Lag (hours)')

plt.tight_layout()
plt.show()

print("ğŸ“Š ACF/PACF Interpretation:")
print("   â€¢ ACF spikes at lag 24  â†’ strong daily seasonality (s=24)")
print("   â€¢ ACF spikes at lag 168 â†’ strong weekly seasonality (s=168)")
print("   â€¢ Slow decay in ACF     â†’ potential non-stationarity (use differencing)")
print("   â€¢ PACF cuts off quickly â†’ AR order typically 1â€“3")
```

## STEP 7 â€” Time Series Decomposition ğŸ”¥

```python

# â”€â”€â”€ TIME SERIES DECOMPOSITION â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# Use 30 days (720 h) for additive decomposition with daily period=24
subset      = ts.iloc[:720]
decomp      = seasonal_decompose(subset, model='additive', period=24, extrapolate_trend='freq')

fig, axes = plt.subplots(4, 1, figsize=(16, 14), sharex=True)
fig.suptitle('Additive Decomposition (First 30 days, period=24h)', fontsize=14, fontweight='bold')

axes[0].plot(subset.index, subset.values,   color='steelblue',  linewidth=0.8)
axes[0].set_title('Original');  axes[0].set_ylabel('MW')

axes[1].plot(decomp.trend.index,    decomp.trend.values,    color='darkgreen',  linewidth=2)
axes[1].set_title('Trend');     axes[1].set_ylabel('Trend MW')

axes[2].plot(decomp.seasonal.index, decomp.seasonal.values, color='darkorange', linewidth=1)
axes[2].set_title('Seasonal (24h)'); axes[2].set_ylabel('Seasonal MW')
axes[2].axhline(0, color='black', linestyle='--', linewidth=0.8, alpha=0.6)

axes[3].plot(decomp.resid.index,    decomp.resid.values,    color='crimson',    linewidth=0.8, alpha=0.8)
axes[3].set_title('Residual'); axes[3].set_ylabel('Residual MW')
axes[3].axhline(0, color='black', linestyle='--', linewidth=1, alpha=0.7)

for ax in axes:
    ax.grid(True, alpha=0.3)
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))

plt.tight_layout()
plt.show()

# Seasonal strength score
var_resid    = np.var(decomp.resid.dropna())
var_seasonal = np.var(decomp.seasonal)
strength_s   = max(0, 1 - var_resid / (var_resid + var_seasonal))

var_trend    = np.var(decomp.trend.dropna())
strength_t   = max(0, 1 - var_resid / (var_resid + var_trend))

print(f"\nğŸ“Š Decomposition Summary:")
print(f"   Seasonal Strength : {strength_s:.4f}  (>0.6 â†’ strong seasonality)")
print(f"   Trend Strength    : {strength_t:.4f}  (>0.6 â†’ strong trend)")
```

## STEP 9 â€” Train / Test Split âœ‚ï¸

We hold out the **last 30 days (720 hours)** as the test set â€” everything before is used for training.


```python

# â”€â”€â”€ TRAIN / TEST SPLIT â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
TEST_HOURS = 24 * 30      # 30 days = 720 hours

ts_full  = df[energy_col].copy()
train_ts = ts_full.iloc[:-TEST_HOURS]
test_ts  = ts_full.iloc[-TEST_HOURS:]

split_date = test_ts.index[0]
print(f"{'='*60}")
print(f"  TRAIN / TEST SPLIT")
print(f"{'='*60}")
print(f"  Train  :  {train_ts.index[0]}  â†’  {train_ts.index[-1]}  ({len(train_ts):,} hrs)")
print(f"  Test   :  {test_ts.index[0]}  â†’  {test_ts.index[-1]}  ({len(test_ts):,} hrs)")
print(f"  Split  :  {split_date}")

# â”€â”€ Visualise split â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
fig, ax = plt.subplots(figsize=(16, 4))
ax.plot(train_ts.index[-24*60:], train_ts.values[-24*60:],
        color='steelblue', linewidth=0.8, label='Train (last 60 days shown)')
ax.plot(test_ts.index, test_ts.values,
        color='crimson',   linewidth=1.2, label=f'Test (30 days)')
ax.axvline(split_date, color='black', linestyle='--', linewidth=2, label='Split point')
ax.set_title('Train â€“ Test Split', fontweight='bold')
ax.set_ylabel('Energy (MW)'); ax.legend(); ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# â”€â”€ Helper: evaluation metrics â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
def evaluate(y_true, y_pred, label="Model"):
    mae  = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    r2   = r2_score(y_true, y_pred)
    print(f"\n  ğŸ“Š {label}")
    print(f"     MAE  : {mae:>10,.2f} MW")
    print(f"     RMSE : {rmse:>10,.2f} MW")
    print(f"     MAPE : {mape:>10,.2f} %")
    print(f"     RÂ²   : {r2:>10,.4f}")
    return {"Model": label, "MAE": round(mae,2), "RMSE": round(rmse,2),
            "MAPE(%)": round(mape,2), "R2": round(r2,4)}

results_log = []   # will store all model results
```

## STEP 10 â€” Baseline Models ğŸ“

Before fancy models, we need a baseline to beat:
- **NaÃ¯ve (Seasonal lag-24h)** â€” predict "same hour yesterday"
- **7-Day Rolling Mean** â€” rolling average of last 168 hours


```python

# â”€â”€â”€ BASELINE MODELS â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

# 1. Seasonal NaÃ¯ve (lag-24h): predict value from 24 hours ago
naive_pred_24  = train_ts.iloc[-TEST_HOURS:].values   \
    if len(train_ts) >= TEST_HOURS                    \
    else ts_full.shift(24).iloc[-TEST_HOURS:].values

# Use shift(24) properly
naive_pred     = ts_full.shift(24).iloc[-TEST_HOURS:]
roll_mean_pred = ts_full.shift(1).rolling(window=168).mean().iloc[-TEST_HOURS:]

# Drop NaN (can't predict first few points)
valid_idx       = naive_pred.dropna().index.intersection(roll_mean_pred.dropna().index)
valid_idx       = valid_idx.intersection(test_ts.index)

y_true_base     = test_ts.loc[valid_idx]
naive_pred_     = naive_pred.loc[valid_idx]
roll_pred_      = roll_mean_pred.loc[valid_idx]

print("BASELINE MODEL EVALUATION")
print("="*60)
r1 = evaluate(y_true_base, naive_pred_, "Seasonal NaÃ¯ve (lag-24h)")
r2 = evaluate(y_true_base, roll_pred_,  "7-Day Rolling Mean (168h)")
results_log.extend([r1, r2])

# â”€â”€ Plot â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
fig, ax = plt.subplots(figsize=(16, 5))
ax.plot(y_true_base.index,  y_true_base.values,   color='black',      linewidth=1.5,  label='Actual')
ax.plot(naive_pred_.index,    naive_pred_.values,    color='steelblue',  linewidth=1,    label='NaÃ¯ve lag-24h', alpha=0.8)
ax.plot(roll_pred_.index,     roll_pred_.values,     color='darkorange', linewidth=1,    label='Rolling Mean 168h', alpha=0.8)
ax.set_title('Baseline Models vs Actual â€” Test Period', fontweight='bold')
ax.set_ylabel('Energy (MW)'); ax.legend(); ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## STEP 11 â€” SARIMA Model ğŸ“ˆ

**SARIMA(1,0,1)(1,1,1)[24]** â€” captures both AR/MA + seasonal (24-hour) components.  
> We train on the last 90 days to keep fitting time manageable.


```python

# â”€â”€â”€ SARIMA â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# Use only last 90 days of training for speed (SARIMA is O(nÂ²))
SARIMA_TRAIN_DAYS = 90
sarima_train = train_ts.iloc[-(24 * SARIMA_TRAIN_DAYS):]
sarima_test  = test_ts.copy()

print(f"Fitting SARIMA(1,0,1)(1,1,1)[24] on {len(sarima_train):,} hours â€¦")

sarima_model = SARIMAX(
    sarima_train,
    order=(1, 0, 1),
    seasonal_order=(1, 1, 1, 24),
    enforce_stationarity=False,
    enforce_invertibility=False
)
sarima_result = sarima_model.fit(disp=False)

# Forecast TEST_HOURS steps ahead
sarima_fc = sarima_result.get_forecast(steps=TEST_HOURS)
sarima_pred  = sarima_fc.predicted_mean
sarima_ci    = sarima_fc.conf_int(alpha=0.05)
sarima_pred.index = sarima_test.index
sarima_ci.index   = sarima_test.index

print(sarima_result.summary().tables[0])

r_sarima = evaluate(sarima_test, sarima_pred, "SARIMA(1,0,1)(1,1,1)[24]")
results_log.append(r_sarima)

# â”€â”€ Plot â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
fig, ax = plt.subplots(figsize=(16, 5))
ax.fill_between(sarima_ci.index, sarima_ci.iloc[:,0], sarima_ci.iloc[:,1],
                alpha=0.2, color='steelblue', label='95% CI')
ax.plot(sarima_test.index, sarima_test.values,  color='black',     linewidth=1.5, label='Actual')
ax.plot(sarima_pred.index, sarima_pred.values,  color='steelblue', linewidth=1.2, label='SARIMA Forecast')
ax.set_title('SARIMA â€” Forecast vs Actual (Test Period)', fontweight='bold')
ax.set_ylabel('Energy (MW)'); ax.legend(); ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## STEP 12 â€” XGBoost Model ğŸŒ³

XGBoost treats time series as a supervised regression problem using lag + calendar features.


```python

# â”€â”€â”€ XGBOOST â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# Select feature columns (drop NaNs from lag creation)
feature_cols = [
    'hour','day','month','year','day_of_week','quarter',
    'is_weekend','is_peak_hour',
    'hour_sin','hour_cos','dow_sin','dow_cos','month_sin','month_cos',
    'lag_1h','lag_24h','lag_48h','lag_168h',
    'roll_mean_24h','roll_std_24h','roll_mean_7d',
    'ewm_mean_24h',
    'diff_1h','diff_24h'
]
# Keep only cols that exist
feature_cols = [c for c in feature_cols if c in df_feat.columns]

df_model = df_feat[[energy_col] + feature_cols].dropna()
X = df_model[feature_cols]
y = df_model[energy_col]

# Time-based split (same split point)
X_train = X.loc[X.index < split_date]
y_train = y.loc[y.index < split_date]
X_test  = X.loc[X.index >= split_date]
y_test  = y.loc[y.index >= split_date]

print(f"XGBoost â€” Train: {X_train.shape}  |  Test: {X_test.shape}")

# â”€â”€ Train â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
xgb_model = xgb.XGBRegressor(
    n_estimators    = 500,
    max_depth       = 6,
    learning_rate   = 0.05,
    subsample       = 0.8,
    colsample_bytree= 0.8,
    min_child_weight= 3,
    reg_alpha       = 0.1,
    reg_lambda      = 1.0,
    random_state    = 42,
    n_jobs          = -1,
    verbosity       = 0
)
xgb_model.fit(X_train, y_train,
              eval_set=[(X_test, y_test)],
              verbose=False)

xgb_pred = pd.Series(xgb_model.predict(X_test), index=X_test.index)

r_xgb = evaluate(y_test, xgb_pred, "XGBoost")
results_log.append(r_xgb)

# â”€â”€ Feature Importance â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
fi = pd.Series(xgb_model.feature_importances_, index=feature_cols).sort_values(ascending=False)

fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# Forecast plot
axes[0].plot(y_test.index,   y_test.values,   color='black',     linewidth=1.2,  label='Actual')
axes[0].plot(xgb_pred.index, xgb_pred.values, color='darkorange', linewidth=1.2, label='XGBoost', alpha=0.9)
axes[0].set_title('XGBoost â€” Forecast vs Actual', fontweight='bold')
axes[0].set_ylabel('Energy (MW)'); axes[0].legend(); axes[0].grid(True, alpha=0.3)

# Feature importance (top 15)
fi.head(15).plot(kind='barh', ax=axes[1], color='teal', edgecolor='white')
axes[1].invert_yaxis()
axes[1].set_title('XGBoost â€” Top 15 Feature Importances', fontweight='bold')
axes[1].set_xlabel('Importance Score')
axes[1].grid(True, alpha=0.3, axis='x')

plt.tight_layout()
plt.show()
```

## STEP 13 â€” Facebook Prophet Model ğŸ”®

Prophet handles seasonality (daily + weekly + yearly) automatically.


```python

# â”€â”€â”€ FACEBOOK PROPHET â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
if not PROPHET_AVAILABLE:
    print("âš ï¸ Prophet not installed. Run:  pip install prophet")
else:
    # Prophet requires column names 'ds' and 'y'
    prophet_train_df = train_ts.reset_index().rename(columns={'Datetime':'ds', energy_col:'y'})
    prophet_test_df  = test_ts.reset_index().rename(columns={'Datetime':'ds', energy_col:'y'})

    # Build model with daily + weekly + yearly seasonality
    prophet_model = Prophet(
        seasonality_mode        = 'additive',
        daily_seasonality       = True,
        weekly_seasonality      = True,
        yearly_seasonality      = True,
        changepoint_prior_scale = 0.05,
        seasonality_prior_scale = 10.0,
        interval_width          = 0.95
    )
    prophet_model.fit(prophet_train_df)

    # Forecast over test period
    future      = prophet_model.make_future_dataframe(periods=TEST_HOURS, freq='h', include_history=False)
    forecast_df = prophet_model.predict(future)

    # Align index
    prophet_pred = forecast_df.set_index('ds')['yhat']
    prophet_pred.index = pd.to_datetime(prophet_pred.index)
    prophet_pred = prophet_pred.loc[prophet_pred.index.isin(test_ts.index)]
    y_true_pr    = test_ts.loc[test_ts.index.isin(prophet_pred.index)]

    r_prophet = evaluate(y_true_pr, prophet_pred, "Prophet")
    results_log.append(r_prophet)

    # â”€â”€ Plot components â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    fig1 = prophet_model.plot_components(forecast_df)
    plt.suptitle('Prophet â€” Component Decomposition', fontsize=13, fontweight='bold', y=1.02)
    plt.tight_layout()
    plt.show()

    # â”€â”€ Forecast vs Actual â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    fig2, ax = plt.subplots(figsize=(16, 5))
    ax.plot(y_true_pr.index,    y_true_pr.values,    color='black',        linewidth=1.5, label='Actual')
    ax.plot(prophet_pred.index, prophet_pred.values, color='mediumorchid', linewidth=1.2, label='Prophet', alpha=0.9)
    ax.set_title('Prophet â€” Forecast vs Actual', fontweight='bold')
    ax.set_ylabel('Energy (MW)'); ax.legend(); ax.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
```

## STEP 14 â€” LSTM (Deep Learning) Model ğŸ§ 

LSTM networks are excellent at learning long-range temporal dependencies from sequences.  
We use a look-back window of **168 hours (1 week)** to predict the next hour.


```python

# â”€â”€â”€ LSTM â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
LOOK_BACK   = 168    # 1 week in hours
BATCH_SIZE  = 64
EPOCHS      = 30

# Scale data
scaler_lstm = MinMaxScaler(feature_range=(0, 1))
ts_scaled   = scaler_lstm.fit_transform(ts_full.values.reshape(-1, 1)).flatten()

# Build sequences
def build_sequences(data, look_back):
    X_seq, y_seq = [], []
    for i in range(look_back, len(data)):
        X_seq.append(data[i - look_back : i])
        y_seq.append(data[i])
    return np.array(X_seq), np.array(y_seq)

# Split index (exclude last TEST_HOURS for test)
split_idx = len(ts_full) - TEST_HOURS

train_scaled = ts_scaled[:split_idx]
test_scaled  = ts_scaled[split_idx - LOOK_BACK:]   # include look-back window

X_tr, y_tr = build_sequences(train_scaled, LOOK_BACK)
X_te, y_te = build_sequences(test_scaled,  LOOK_BACK)

# Reshape: (samples, timesteps, features)
X_tr = X_tr.reshape(*X_tr.shape, 1)
X_te = X_te.reshape(*X_te.shape, 1)

print(f"LSTM input shapes â€” Train: {X_tr.shape}  |  Test: {X_te.shape}")

# â”€â”€ Build model â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
tf.random.set_seed(42)
lstm_model = Sequential([
    LSTM(128, return_sequences=True,  input_shape=(LOOK_BACK, 1)),
    Dropout(0.2),
    LSTM(64,  return_sequences=False),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dense(1)
], name="LSTM_Energy_Forecast")
lstm_model.summary()

lstm_model.compile(optimizer='adam', loss='mse', metrics=['mae'])

callbacks = [
    EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True, verbose=1),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6, verbose=1)
]

history = lstm_model.fit(
    X_tr, y_tr,
    validation_split = 0.1,
    epochs           = EPOCHS,
    batch_size       = BATCH_SIZE,
    callbacks        = callbacks,
    verbose          = 1
)

# â”€â”€ Predict & inverse-scale â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
lstm_pred_scaled = lstm_model.predict(X_te, verbose=0).flatten()
lstm_pred        = scaler_lstm.inverse_transform(lstm_pred_scaled.reshape(-1,1)).flatten()
y_true_lstm      = test_ts.values[:len(lstm_pred)]
lstm_pred_series = pd.Series(lstm_pred, index=test_ts.index[:len(lstm_pred)])

r_lstm = evaluate(y_true_lstm, lstm_pred, "LSTM (168h look-back)")
results_log.append(r_lstm)

# â”€â”€ Training loss plot â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
fig, axes = plt.subplots(1, 2, figsize=(18, 5))

axes[0].plot(history.history['loss'],     label='Train Loss', color='steelblue')
axes[0].plot(history.history['val_loss'], label='Val Loss',   color='crimson')
axes[0].set_title('LSTM â€” Training History', fontweight='bold')
axes[0].set_xlabel('Epoch'); axes[0].set_ylabel('MSE Loss')
axes[0].legend(); axes[0].grid(True, alpha=0.3)

axes[1].plot(test_ts.index[:len(lstm_pred)], y_true_lstm,   color='black',     linewidth=1.5, label='Actual')
axes[1].plot(lstm_pred_series.index,         lstm_pred,     color='royalblue', linewidth=1.2, label='LSTM', alpha=0.9)
axes[1].set_title('LSTM â€” Forecast vs Actual', fontweight='bold')
axes[1].set_ylabel('Energy (MW)'); axes[1].legend(); axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
