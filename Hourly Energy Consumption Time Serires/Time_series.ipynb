```python
# Visualize rolling statistics
def plot_rolling_statistics(series, window=24*30):
    """
    Plot rolling mean and standard deviation
    """
    rolling_mean = series.rolling(window=window).mean()
    rolling_std = series.rolling(window=window).std()
    
    plt.figure(figsize=(15, 6))
    plt.plot(series, color='blue', label='Original', alpha=0.5)
    plt.plot(rolling_mean, color='red', label=f'Rolling Mean (window={window})', linewidth=2)
    plt.plot(rolling_std, color='green', label=f'Rolling Std (window={window})', linewidth=2)
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation - PJME', fontsize=14, fontweight='bold')
    plt.xlabel('Date')
    plt.ylabel('Power (MW)')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

plot_rolling_statistics(data['PJME']['PJME_MW'])
```

```python
from statsmodels.tsa.stattools import adfuller

# Perform Augmented Dickey-Fuller test
def adf_test(series, title=''):
    """
    Perform Augmented Dickey-Fuller test for stationarity
    """
    print(f'Augmented Dickey-Fuller Test: {title}')
    print('='*60)
    
    result = adfuller(series.dropna(), autolag='AIC')
    
    labels = ['ADF Test Statistic', 'p-value', '# Lags Used', '# Observations']
    output = pd.Series(result[0:4], index=labels)
    
    for key, value in result[4].items():
        output[f'Critical Value ({key})'] = value
    
    print(output.to_string())
    
    if result[1] <= 0.05:
        print("\n‚úÖ Result: Series is STATIONARY (p-value <= 0.05)")
    else:
        print("\n‚ùå Result: Series is NON-STATIONARY (p-value > 0.05)")
    print('='*60)
    
# Test on original PJME data
adf_test(data['PJME']['PJME_MW'], 'PJME Energy Consumption (Original)')
```

```python
# Fit ARIMA model
print("Fitting ARIMA(1,1,1) model...")
model = ARIMA(train, order=(1, 1, 1))
model_fit = model.fit()

# Display model summary
print(model_fit.summary())

# Make predictions
forecast = model_fit.forecast(steps=len(test))

# Calculate error metrics
mae = mean_absolute_error(test, forecast)
rmse = np.sqrt(mean_squared_error(test, forecast))
mape = np.mean(np.abs((test - forecast) / test)) * 100

print("\n" + "="*60)
print("Model Performance Metrics:")
print("="*60)
print(f"MAE (Mean Absolute Error): {mae:.2f} MW")
print(f"RMSE (Root Mean Squared Error): {rmse:.2f} MW")
print(f"MAPE (Mean Absolute Percentage Error): {mape:.2f}%")
print("="*60)

# Visualization
plt.figure(figsize=(15, 6))
plt.plot(train.index, train, label='Training Data', color='blue', alpha=0.6)
plt.plot(test.index, test, label='Actual Test Data', color='green', linewidth=2)
plt.plot(test.index, forecast, label='ARIMA Forecast', color='red', linestyle='--', linewidth=2)
plt.title('ARIMA Model Forecast vs Actual (PJME)', fontsize=14, fontweight='bold')
plt.xlabel('Time')
plt.ylabel('Power (MW)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Residual Analysis
residuals = model_fit.resid

fig, axes = plt.subplots(2, 2, figsize=(16, 10))

# Residuals over time
axes[0, 0].plot(residuals)
axes[0, 0].set_title('Residuals Over Time', fontsize=12, fontweight='bold')
axes[0, 0].set_xlabel('Time')
axes[0, 0].set_ylabel('Residuals')
axes[0, 0].axhline(y=0, color='r', linestyle='--')
axes[0, 0].grid(True, alpha=0.3)

# Histogram of residuals
axes[0, 1].hist(residuals, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
axes[0, 1].set_title('Residuals Distribution', fontsize=12, fontweight='bold')
axes[0, 1].set_xlabel('Residuals')
axes[0, 1].set_ylabel('Frequency')
axes[0, 1].grid(True, alpha=0.3)

# Q-Q plot
from scipy import stats
stats.probplot(residuals, dist="norm", plot=axes[1, 0])
axes[1, 0].set_title('Q-Q Plot', fontsize=12, fontweight='bold')
axes[1, 0].grid(True, alpha=0.3)

# ACF of residuals
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(residuals, lags=40, ax=axes[1, 1])
axes[1, 1].set_title('ACF of Residuals', fontsize=12, fontweight='bold')
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Statistical tests on residuals
print("Residual Statistics:")
print("="*60)
print(f"Mean of residuals: {residuals.mean():.4f}")
print(f"Std of residuals: {residuals.std():.4f}")
print("="*60)
```
